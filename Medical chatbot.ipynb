```python
# Define the medical chatbot prompt template
prompt_template = """You are an AI medical assistant. Your task is to analyze the given medical query and determine the most appropriate medical department the user should consult.

### Task:
- If the question is about **skin-related** issues, return **Dermatology**.
- If the question is about **bones or joints**, return **Orthopedics**.
- If the issue is **related to the heart**, return **Cardiology**.
- If the problem involves the **nervous system (brain, spinal cord, nerves)**, return **Neurology**.
- If the concern is about the **stomach, digestion, or liver**, return **Gastroenterology**.
- If the problem involves **hormones or diabetes**, return **Endocrinology**.
- If it is about **general symptoms like fever, cold, or common body aches**, return **General Medicine**.
- If the query is **not related to a known medical specialty**, return **No department for this disease**.

### User Query:
{}

### Recommended Department:
"""

# Function to get the department based on user query
def get_medical_department(user_query):
    # Ensure prompt_template is correctly formatted with the user query
    prompt = prompt_template.format(user_query)

    # Tokenize input and generate response
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512)
    outputs = model.generate(**inputs, max_new_tokens=50)

    # Decode output to get the response text
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract the department name from the response
    department = response.split("### Recommended Department:")[-1].strip()

    return department

# Example usage
while True:
    query = input("Enter your medical query (or type 'exit' to stop): ")
    if query.lower() == "exit":
        break
    department = get_medical_department(query)
    print(f"Recommended Department: {department}")
```

```python
from datetime import datetime

def convert_to_24hr(time_str):
    """Convert '10 AM' to '10:00' and '2 PM' to '14:00'"""
    return datetime.strptime(time_str.strip(), "%I %p").strftime("%H:%M")

def fix_time_format(time_range):
    """Convert '10 AM - 2 PM' to '10:00 - 14:00'"""
    start, end = time_range.split("-")
    return f"{convert_to_24hr(start)} - {convert_to_24hr(end)}"

# Apply to dataset
doctors_df["availability"] = doctors_df["availability"].apply(fix_time_format)
```
